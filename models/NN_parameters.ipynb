```python
from sklearn.externals import joblib
import matplotlib.pyplot as plt
import math
import numpy as np
%matplotlib inline
```

```python
H = joblib.load('NN_hidden_results.pkl')
```

```python
for key in ['rmse', 'mae']:
    values = []
    for i in range(len(H)):
        values.append(H[i][key][0])
    plt.plot(range(1, len(H)+1), values, label=key.upper())
    values = np.array(values)
    print key.upper(), 'best n_hidden:', np.argmin(values), 'value:', np.min(values)
    print values.tolist()
plt.legend()
plt.xlabel('# hidden units')
plt.show()
```

```python
for key in H[0].keys():
    values = []
    for i in range(len(H)):
        values.append(H[i][key][0])
    values = filter(lambda x: not math.isnan(x), values)
    plt.figure()
    plt.plot(range(1, len(values)+1), values, label=key.upper())
    plt.legend()
    plt.xlabel('# hidden units')
    plt.show()
    values = np.array(values)
    argcmp = np.argmin
    cmp = np.min
    if key in ['corr', 'r2']:
        cmp = np.max
        argcmp = np.argmax
    print key.upper(), 'best n_hidden:', argcmp(values), 'value:', cmp(values)
    print values.tolist()
```
